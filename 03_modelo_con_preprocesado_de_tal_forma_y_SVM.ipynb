Dividimos el conjunto de datos en características (X) y etiquetas (y). La etiqueta a predecir es `RENDIMIENTO_GLOBAL`, una variable categórica que representa el nivel de rendimiento de los estudiantes.


```python
X = d_p.drop(columns='RENDIMIENTO_GLOBAL').values
y = d_p['RENDIMIENTO_GLOBAL'].values
```

Usamos funciones auxiliares para dividir los datos, entrenar y evaluar el modelo.

```python
def split_data(X, y, pct):
    assert len(X) == len(y), "X and y must have the same length"
    n = len(X)
    indices = np.random.permutation(n)
    train_size = int(n * pct)
    train_idx = indices[:train_size]
    test_idx = indices[train_size:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)

def split_fit_predict(estimator, X, y, train_pct):
    Xtr, Xts, ytr, yts = split_data(X, y, train_pct)
    estimator.fit(Xtr, ytr)
    preds_ts = estimator.predict(Xts)
    return estimator, Xts, yts, accuracy(yts, preds_ts)
```

## 6. Entrenamiento del modelo SVM


```python
# Usamos un subconjunto pequeño (10%) para entrenamiento más rápido
X_train, X_test, y_train, y_test = split_data(X, y, 0.1)

# Inicializar modelo LinearSVC
modelo_svm = LinearSVC(C=1.0, max_iter=4000, random_state=42)
```

## Entrenamiento del modelo SVM

En esta celda entrenamos un modelo de Máquina de Vectores de Soporte (SVM) lineal utilizando `LinearSVC` de Scikit-learn.  
La línea `modelo_svm.fit(X_train, y_train)` ajusta el modelo a los datos de entrenamiento.

Este proceso consiste en encontrar un hiperplano que separe lo mejor posible las clases del conjunto de datos, maximizando el margen entre las clases distintas.

Es importante que los datos estén previamente escalados, ya que los SVM son sensibles a la escala de las variables.  
Una vez entrenado, el modelo podrá ser usado para hacer predicciones con datos nuevos utilizando el método `.predict()`.


```python
modelo_svm.fit(X_train, y_train)
```

## 7. Evaluación del modelo

```python
predicciones = modelo_svm.predict(X_test)

# Calcular y mostrar precisión
precision = accuracy_score(y_test, predicciones)
print(f"Precisión en subconjunto de prueba: {precision:.4f}")
```
