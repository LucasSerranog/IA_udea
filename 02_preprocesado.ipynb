Hacemos la columna **"id"** como indice de nuestro conjunto de datos

```python
if 'index' in d.columns:
    d.drop(columns=['index'], inplace=True)

# Establecer 'ID' como índice del DataFrame
d.set_index('ID', inplace=True)

# Verificar el resultado
d.head()
```

```python
d.ESTU_PRIVADO_LIBERTAD = d.ESTU_PRIVADO_LIBERTAD.fillna('no info')
d.ESTU_PRIVADO_LIBERTAD.value_counts()
```

Hacemos el Conteo de estudiantes totales de la encuesta

```python
def contar_estudiantes(df):
    """
    Retorna la cantidad de estudiantes únicos en el DataFrame.
    Asume que cada estudiante tiene un ID único como índice.
    """
    return df.shape[0]
cantidad = contar_estudiantes(d)
print(f"Cantidad de estudiantes: {cantidad}")
```

limpiamos datos faltantes y luego analizar la frecuencia de respuestas

```python
d.FAMI_TIENEINTERNET = d.FAMI_TIENEINTERNET.fillna('no info')
d.FAMI_TIENEINTERNET.value_counts()
```

```python
# Extraer la columna original sin convertir a string
x = d.FAMI_TIENEINTERNET.values

# Obtener solo los valores únicos válidos (no NaN)
FAMI_TIENEINTERNET_vals = sorted(pd.Series(x).dropna().unique())

# Crear diccionario para codificación one-hot
FAMI_TIENEINTERNET_onehot_vals = {
    val: np.eye(len(FAMI_TIENEINTERNET_vals))[i] for i, val in enumerate(FAMI_TIENEINTERNET_vals)
}

# Crear codificación, respetando los NaN
FAMI_TIENEINTERNET_onehot_enc = np.full((len(x), len(FAMI_TIENEINTERNET_vals)), np.nan)
for i, val in enumerate(x):
    if pd.notna(val):
        FAMI_TIENEINTERNET_onehot_enc[i] = FAMI_TIENEINTERNET_onehot_vals[val]

# Crear DataFrame con columnas
FAMI_TIENEINTERNET_df = pd.DataFrame(
    FAMI_TIENEINTERNET_onehot_enc,
    columns=[f"FAMI_TIENEINTERNET__{v}" for v in FAMI_TIENEINTERNET_vals],
    index=d.index
)

# Concatenar al dataframe original
d = pd.concat([FAMI_TIENEINTERNET_df, d.drop(columns=["FAMI_TIENEINTERNET"])], axis=1)

```
